```python
in_path = 'D:/LIVERLAST/dicom_file/images'
out_path = 'D:/LIVERLAST/dicom_groups/images'
```

```python
from glob import glob
import shutil
import os
import dicom2nifti
import nibabel as nib
import numpy as np
```

```python
patient_list = glob(in_path + '/*')
print(patient_list[0])
```

```python
for patient in glob(in_path + '/*'):
    patient_name = os.path.basename(os.path.normpath(patient))
    number_folders = int(len(glob(patient + '/*'))/64)

    for i in range(number_folders):
        output_path_name = os.path.join(out_path, patient_name+ '_' + str(i))
        os.mkdir(output_path_name)
        for i, file in enumerate(glob(patient+'/*')):
            if i == 64 + 1:
                break
            shutil.move(file, output_path_name)
```

```python
in_path_images = 'D:/LIVERLAST/dicom_groups/images/*'
in_path_labels = 'D:/LIVERLAST/dicom_groups/labels/*'
out_path_images = 'D:/LIVERLAST/nifti_files/images'
out_path_labels = 'D:/LIVERLAST/nifti_files/labels'
list_images = glob(in_path_images)
list_labels = glob(in_path_images)
```

```python
for patient in list_labels:
    patient_name = os.path.basename(os.path.normpath(patient))
    dicom2nifti.dicom_series_to_nifti(patient, os.path.join(out_path_labels, patient_name + '.nii.gz'))
```

```python
input_nifti_file_path = 'D:/LIVERLAST/nifti_files/labels/*'
list_labels = glob(input_nifti_file_path)

for patient in list_labels:
    nifti_file = nib.load(patient)
    fdata = nifti_file.get_fdata()
    np_unique = np.unique(fdata)
    if len(np_unique) == 1:
        print(patient)
```
